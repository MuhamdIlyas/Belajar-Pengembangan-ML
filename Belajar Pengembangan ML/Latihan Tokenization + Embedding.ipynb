```python
#menggunakan tokinizer
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
#membuat objek tokenizer dengan memanggil fungsi tokenizer dan melengkapi parameternya
tokenizer = Tokenizer(num_words=15, oov_token='-')
```

```python
#membuat teks yang ditokenisasi dan dipakai untuk pelatihan model
teks = ['Saya Suka Programming',
        'Programming Sangat Menyenangkan',
        'Machine Learning Berbeda dengan Pemrograman Konvensional']
```

```python
#melakukan tokenisasi
tokenizer.fit_on_texts(teks)
```

```python
#mengubah kalimat kedalam nilai
sequences = tokenizer.texts_to_sequences(teks)
```

```python
#melihat hasil tokenisasi
print(tokenizer.word_index)
```

```python
#contoh kalimat lain yang (Kata ‘belajar’, ‘sejak’, dan ‘SMP’ tidak ada memiliki token pada dictionary hasil tokenisasi)
print(tokenizer.texts_to_sequences(['Saya Suka Programming']))
print(tokenizer.texts_to_sequences(['Saya Suka Belajar Programming sejak SMP']))
```

```python
#menambahkan fungsi padding(membuat setiap kalimat pada teks memiliki panjang yang seragam)
from tensorflow.keras.preprocessing.sequence import pad_sequences
sequences_samapanjang = pad_sequences(sequences)
```

```python
#melihat hasil paddin
print(sequences_samapanjang)
```

```python
#merubah sehingga 0 ditambahkan di akhir sequence
sequences_samapanjang = pad_sequences(sequences,
                                      padding='post',
                                      maxlen=5)
```

```python
#melihat hasil paddin
print(sequences_samapanjang)
```

```python
#mengubah pengaturan dan mengambil 5 kata terakhir dari tiap kalimat
sequences_samapanjang = pad_sequences(sequences,
                                      padding='post',
                                      maxlen=5,
                                      truncating='post')
```

```python
#melihat hasil paddin
print(sequences_samapanjang)
```
