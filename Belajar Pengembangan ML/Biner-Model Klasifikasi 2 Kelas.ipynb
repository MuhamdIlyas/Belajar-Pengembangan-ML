```python
#memanggil file dengan pandas
import pandas as pd

path = '/content/drive/MyDrive/Dataset/citrus.csv'
df = pd.read_csv(path)

#melihat overview
df.info()
```

```python
#menampilkan dataframe
df.head()
```

```python
#mengubah nilai-nilai pada kolom name menjadi bilangan numerik
df.name[df.name == 'orange'] = 0
df.name[df.name == 'grapefruit'] = 1
```

```python
#Values mengembalikan numpy array yang dikonversi dari dataframe
dataset = df.values

#Untuk melihat dataframe yang telah diubah menjadi array
dataset
```

```python
#memisahkan atribut dan label
#pilih 4 kolom terakhir sebagai atribut
X = dataset[:, 1:6]
# bilangan sebelum koma untuk memilih baris pada dataframe
# bilangan setelah koma untuk memilih kolom pada dataframe
```

```python
#membuat array numpy yang berisi label saja
y = dataset[:, 0]
```

```python
#melakukan normalisasi data
from sklearn import  preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

```python
#memisahkan data training dan testing
from sklearn.model_selection import  train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

```python
#Karena tadi kita mengubah nilai-nilai pada kolom name menjadi bilangan numerik, 
#yang artinya kita mengubah label menjadi tipe data boolean

import numpy as np
Y_train = Y_train.astype(np.float32)
Y_test = Y_test.astype(np.float32)
```

```python
#Membangun model jst
from keras.models import Sequential
from keras.layers import Dense
```

```python
#model sequential yang memiliki 3 buah layer
model = Sequential ([
                     Dense(32, activation='relu', input_shape=(5,)),
                     Dense(32, activation='relu'),
                     Dense(1, activation='sigmoid'),])
```

```python
#menentukan optimizer dan loss
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
#melatih model
model.fit(X_train, Y_train, epochs=100)
```

```python
#melihat nilai loss dan akuras
model.fit(X_test, Y_test)
```
