```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Dataset/sentiment labelled sentences/yelp_labelled.txt', names=['sentence', 'label'], sep='\t')
```

```python
#melihat 5 sampel terakhir dari dataset
df.tail()
```

```python
#membagi dataset menjadi train dan test set
from sklearn.model_selection import train_test_split
kalimat = df['sentence'].values
y = df['label'].values
kalimat_latih, kalimat_test, y_latih, y_test = train_test_split(kalimat, y, test_size=0.2)
```

```python
#menggunakan tokenisasi
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=250, oov_token='x')
tokenizer.fit_on_texts(kalimat_latih)
tokenizer.fit_on_texts(kalimat_test)

sequens_latih = tokenizer.texts_to_sequences(kalimat_latih)
sequens_test = tokenizer.texts_to_sequences(kalimat_test)

padded_latih = pad_sequences(sequens_latih)
padded_test = pad_sequences(sequens_latih)
```

```python
#arsitektur sequential
import tensorflow as tf
model = tf.keras.Sequential([
      tf.keras.layers.Embedding(250, 16, input_length=20),
      tf.keras.layers.GlobalAveragePooling1D(),
      tf.keras.layers.Dense(24, activation='relu'),
      tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(
    loss='binary_crossentropy', 
    optimizer='adam', 
    metrics=['accuracy'])
```

```python
#menentukan loss function dan optimizer
num_epochs = 30
history = model.fit(padded_latih, y_latih, epochs=num_epochs,
                    validation_data=(padded_test, y_test),
                    verbose=2)
```
